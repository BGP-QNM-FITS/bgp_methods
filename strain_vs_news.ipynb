```python
import pickle 
import json
import matplotlib.pyplot as plt
import numpy as np
import qnmfits 
import bgp_qnm_fits as bgp 
import CCE as CCE 
```

```python
filepath  = '/data/rvnd2-2/CCE_data'
```

```python
filenum = '0001'
rssname = 'R2'
lev_upper = "Lev5"
lev_lower = "Lev4"

START_TIME = -10
END_TIME = 260
TIME_STEP = 0.1

times = np.arange(START_TIME, START_TIME + END_TIME, TIME_STEP)
```

```python
sim_main = CCE.SXS_CCE(filenum, lev="Lev5", radius="R2")
sim_lower = CCE.SXS_CCE(filenum, lev="Lev4", radius="R2")
```

```python
h_residual = bgp.get_residuals(sim_main, sim_lower, START_TIME, END_TIME, TIME_STEP)
```

```python
sim_main_news = CCE.SXS_CCE(filenum, lev="Lev5", radius="R2", type="news")
sim_lower_news = CCE.SXS_CCE(filenum, lev="Lev4", radius="R2", type="news")
```

```python
h_residual_news = bgp.get_residuals(sim_main_news, sim_lower_news, START_TIME, END_TIME, TIME_STEP)
```

```python
plt.plot(h_residual['times'], np.abs(h_residual['h'][(2,2)].real), label='h residuals')
plt.plot(h_residual_news['times'], np.abs(h_residual_news['h'][(2,2)].real), label='h residuals news')
plt.yscale('log')
plt.legend() 
```
